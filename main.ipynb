```python
import pandas as pd
```

```python
#all of high school GPAs from the Common Data Sets provided by all Big Ten Universities as a dataset
df = pd.read_csv("https://waf.cs.illinois.edu/discovery/cds-high-school-gpas.csv")
```

```python
def best_n(df, N):
    ''' finds the N rows that have the highest percentage of incoming freshmen with a 4.00 GPA
    returns it as a dataframe'''
    upper = 100
    lower = 0
    while upper - lower > 0.01:
        mid = (upper + lower)/2
        if len(df[df["=4.00"] > mid]) < N:
            upper = mid
        elif len(df[df["=4.00"] > mid]) > N:
            lower = mid
        else:
            return df[df["=4.00"] > mid].reset_index(drop = True)
    return 0   
```

```python
df_UWM = df[df["School"] == "University of Wisconsin-Madison"].reset_index(drop = True)
df_UM = df[df["School"] == "University of Michigan"].reset_index(drop = True)
df_UMS = df[df["School"] == "Michigan State"].reset_index(drop = True)
```

```python
frame1 = df_UWM.plot.line(x = 'Year (Fall)',
                         y = '>=3.75',
                         title = 'trends in HS GPA',
                         ylabel = r"% of students with GPS >= 3.75")

df_UM.plot.line(x = 'Year (Fall)',
                         y = '>=3.75',
                         title = 'trends in HS GPA',
                         ylabel = r"% of students with GPS >= 3.75",
                         ax = frame1)

df_UMS.plot.line(x = 'Year (Fall)',
                         y = '>=3.75',
                         title = 'trends in HS GPA',
                         ylabel = r"% of students with GPS >= 3.75",
                         ax = frame1)
frame1.legend( ["Wisconsin", "Michigan", "Michigan State"])
frame1.set_xticks(range(2004,2024))
frame1.set_xticklabels(labels=range(2004, 2024), rotation=90) #rotates the xlabels by 90 degrees
frame1.set_ylim(0,100)
```
